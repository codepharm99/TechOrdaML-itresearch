```python
import csv
import json

data_rows = []   # Мұнда алғашқы 10 жол сақталады

with open("Salaries.csv", encoding="utf-8") as file:
    reader = csv.reader(file)
    header = next(reader)  # Тақырып жолын өткізіп кетеміз
    for i, row in enumerate(reader):
        if i >= 10:   # Тек алғашқы 10 дерек
            break
        data_rows.append(row)

print("Барлығы:", len(data_rows), "жол")
print("Бірінші қызметкер дерегі:\n", data_rows[0])
print("Бағандар:", header)
```

```python
name_idx = header.index("EmployeeName")
title_idx = header.index("JobTitle")
first_three_names = tuple(row[name_idx] for row in data_rows[:3])
employee_titles = {row[name_idx]: row[title_idx] for row in data_rows[:3]}
print("Кортеж (3 адам):", first_three_names)
print("\nСөздік (аты-жөні → қызмет атағы):")
print(json.dumps(employee_titles, indent=4, ensure_ascii=False))
```

```python
basepay_idx = header.index("BasePay")

base_pays_str = [row[basepay_idx] for row in data_rows]

print("Базалық жалақылар (мәтіндік түрде):")
print(base_pays_str)
```

```python
job_titles = []
for row in data_rows:
    job_titles.append(row[title_idx])
unique_titles = set(job_titles)
print("Барлық бірегей қызмет атақтары:")
for title in unique_titles:
    print("-", title)
print("\nжиын пайдалы, себебі ол қайталанатын мәндерді автоматты түрде алып тастайды")
print("бұл деректердің тазалығын сақтауға және жылдам іздеуге мүмкіндік беред")
```

```python
results = {
    "first_three_names": first_three_names,
    "employee_titles": employee_titles,
    "base_pays_str": base_pays_str,
    "unique_titles": list(unique_titles)
}

with open("results_TT3.json", "w", encoding="utf-8") as f:
    json.dump(results, f, ensure_ascii=False, indent=4)

print("results_TT3.json")
```
