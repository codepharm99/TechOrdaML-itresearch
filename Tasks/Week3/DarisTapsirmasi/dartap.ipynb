```python
students = {
    "Айжан": 85,
    "Мұрат": 92,
    "Жанель": 78,
    "Дамир": 95,
    "Аружан": 88
}
max_score = max(students.values())
best_students = [name for name, score in students.items() if score == max_score]
print("Ең жоғары баға:", max_score)
print("Бұл бағаны алған студент(тер):", ", ".join(best_students))
```
